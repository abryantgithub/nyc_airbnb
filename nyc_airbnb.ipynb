```python
print("\nMissing Values")
print(df.isnull().sum())
```

```python
print("\nDistribution of Numeric Features")
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
print("\nDistribution of Price")
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=100, kde=True)
plt.title('Distribution of Price')
plt.show()
```

```python
print("\nBoxplot of Price by Room Type")
plt.figure(figsize=(10,6))
sns.boxplot(x='room_type', y='price',
            data=df)
plt.title('Boxplot of Price by Room Type')
plt.show()
```

```python
print("\nCountplot of Neighbourhood Group")
plt.figure(figsize=(10,6))
sns.countplot(y='neighbourhood_group', data=df, palette='Set1')
plt.title('Count of Listings by Neighbourhood Group')
plt.show()
```

```python
# After converting to numeric
df_filtered = df[pd.to_numeric(df['number_of_reviews'], errors='coerce').notna()]

print("\nScatterplot: Price vs. Number of Reviews")
plt.figure(figsize=(10,6))
sns.scatterplot(x='number_of_reviews', y='price', data=df_filtered)
plt.title('Price vs. Number of Reviews')
plt.show()
```
